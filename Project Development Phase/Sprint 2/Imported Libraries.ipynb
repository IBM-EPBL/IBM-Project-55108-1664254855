```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
x_train = train_datagen.flow_from_directory(r'/content/drive/MyDrive/TRAIN_SET',target_size=(64,64),batch_size=32,class_mode='categorical')
x_test = test_datagen.flow_from_directory(r'/content/drive/MyDrive/TEST_SET',target_size=(64,64),batch_size=32,class_mode='categorical')
```

```python
x_train.class_indices
```

```python
model = Sequential()
```

```python
model.add(Convolution2D(32,(3,3),input_shape = (64,64,3),activation ="relu"))
```

```python

model.add(MaxPooling2D(pool_size =(2,2)))
```

```python
model.add(Flatten())
```

```python

model.add(Dense(units = 128 ,kernel_initializer ="uniform" , activation = "relu"))
```

```python
model.add(Dense(units = 5 ,kernel_initializer ="uniform" , activation = "softmax"))
```

```python
model.compile(optimizer = "adam", loss="categorical_crossentropy",metrics = ["accuracy"])
```

```python
model.fit_generator(x_train,steps_per_epoch=47,epochs=10,validation_data=x_test ,validation_steps =20)
```

```python
model.save("fruit.h5")
```
